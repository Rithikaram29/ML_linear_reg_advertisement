```python
import sys
sys.path.append("../")
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from src.get_data import load_data
from src.preprocessor import split_data
from src.model import model_prediction

from sklearn.metrics import r2_score

sc = StandardScaler()
le = LabelEncoder()

df = load_data("../data/Dummy Data HSS.csv")

df.head()

# df.describe()

# df.isnull().sum()
numeric_columns = df.select_dtypes(include="number").columns

df['Influencer'] = le.fit_transform(df['Influencer'])
# df
```

```python
# df.median(numeric_only=True)

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

target_colum = df["Sales"]
# target_colum

independant_col = df.drop(columns=["Sales"])
# print(independant_col)
```

```python
# Preprocessing
x_train, x_test, y_train, y_test = split_data(independant_col, target_colum)

# Reshaping the test
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
# x_train

results = model_prediction(x_train=x_train,y_train=y_train,x_test=x_test)
# print("type:", type(results))

r2 = r2_score(y_test,results)
r2
```
